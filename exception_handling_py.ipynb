```python
try:
  age=int(input("please enter your age:"))
  if age == 18:
    print("your age is a valid age for voting")
  else:
    print("your age is not a valid age for voting")
except:
  if age%2 == 0:
    print("your age is a n even number")
  else:
    print("your age is an odd number")
```
