```python
import math
number=float(input("enter a number"))
if number<0:
  print("square root of an negative number can't be calculated")
else:
  result=math.sqrt(number)
  print(f"the square root result of {number} is {result}")
print("here is the square root answer of your given number:", number)
print("and here is the answer of your number:",result)
```
